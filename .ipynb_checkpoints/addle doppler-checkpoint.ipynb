```python
mse_akorn = []
mse_addle = []
mse_addle_1 = []
mse_tf = []
mse_tf_dof = []
# Ts = [100, 5000, 100_000, 200_000, 500_000]
# Ts = [50, 100, 200, 500, 1000, 2500, 5000]
# Ts = [2500, 5000, 10_000, 20_000]
Ts = [50, 100]



# # Ts = [100]
runs_per_T = 10

from util import *

akorn_fv = None
dof_fv = None
tf_fv = None
addle_fv = None
    
for T in Ts:
    print("T = ", T)
    X = np.linspace(0, T, T)/T 


    G = doppler(abs(1 - X), 0.38)

    mses_T_tf = []
    mses_T_tf_dof = []
    mses_T_akorn = []
    mses_T_addle = []
    mses_T_addle_1 = []

    for l in range(runs_per_T):
        print("Run number " + str(l))
        Y = (G + np.random.normal(0, std, T))
        online = FLH(X, Y, T, lr = 1/2*(np.max(np.abs(Y))**2))
        online.run()
        fv_addle = online.predictions
        
        online_1 = FLH(X, Y, T, lr = 1/8*(np.max(np.abs(Y))**2))
        online_1.run()
        fv_addle_1 = online_1.predictions
                
        mses_T_addle.append(((fv_addle - G)**2).mean())
        mses_T_addle_1.append(((fv_addle_1 - G)**2).mean())

        addle_fv = fv_addle
        addle_fv_1 = fv_addle_1


    #     Z_tf.append(best_tf_fit)

    mse_addle.append(mses_T_addle)
    mse_addle_1.append(mses_T_addle_1)

    


avg_mses_addle = [np.array(mse_T).mean() for mse_T in mse_addle]
avg_mses_addle_1 = [np.array(mse_T).mean() for mse_T in mse_addle_1]

print("Done!")
```

```python
from sklearn.linear_model import LinearRegression
addle_ref = LinearRegression().fit(np.log(np.array(Ts)).reshape(-1, 1), np.log(avg_mses_addle))
```
